```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read snml lengths file
dim50 = np.genfromtxt('31epochs/50dim/scope-100-snml_length.txt', delimiter=',').astype(float)
dim100 = np.genfromtxt('31epochs/100dim/scope-100-snml_length.txt', delimiter=',').astype(float)
dim150 = np.genfromtxt('31epochs/150dim/scope-100-snml_length.txt', delimiter=',').astype(float)
dim200 = np.genfromtxt('31epochs/200dim/scope-100-snml_length.txt', delimiter=',').astype(float)
```

```python
print('50dim: ', sum(dim50))
print('100dim: ', sum(dim100))
print('150dim: ', sum(dim150))
print('200dim: ', sum(dim200))
```

```python
cum_dim50 = np.cumsum(np.flip(dim50)
cum_dim100 = np.cumsum(np.flip(dim100)
cum_dim150 = np.cumsum(dim150)
cum_dim200 = np.cumsum(dim200)

plt.figure(figsize=(15,10))
plt.grid()
plt.plot(cum_dim50[80:], 'r', label='50 dim')
plt.plot(cum_dim100[80:], 'b', label='100 dim')
plt.plot(cum_dim150[80:], 'g', label='150 dim')
plt.plot(cum_dim200[80:], 'y', label='200 dim')
plt.legend()
plt.show()
```
